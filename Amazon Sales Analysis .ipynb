## Exploratory Data Analysis

```python
ax = sns.countplot(x="Size",data=df)

for bar in ax.containers:
    ax.bar_label(bar)
```

#### Note: The graph indicates that M-size is the most commonly purchased size among customers.

```python
df.groupby(["Size"], as_index=False)["Qty"] \
  .sum() \
  .sort_values(by="Qty", ascending=False)
```

```python
s_Qty = (
    df.groupby("Size", as_index=False)["Qty"]
      .sum()
      .sort_values(by="Qty", ascending=False)
)

sns.barplot(x="Size", y="Qty", data=s_Qty)
```

#### Note: The graph indicates that M-size is the most frequently purchased size in terms of quantity sold.

```python
sns.countplot(data=df,x= "Courier Status",hue= "Status")
```

#### Note: From the above graph, the majority of orders are shipped through the courier

```python
plt.figure(figsize=(10, 5))
plt.hist(df["Category"], bins=10, edgecolor="black")
plt.show()
```

#### Note:The graph indicates that T-shirts have the highest number of buyers compared to other product categories

```python
B2B_Check = df["B2B"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(B2B_Check, labels=B2B_Check.index, autopct="%1.1f%%")
plt.title("B2B Orders Distribution")
plt.show()
```

#### Note: From the above chart, it is evident that the vast majority of buyers (99.3%) are retailers, while only 0.7% are B2B buyers.

```python
plt.figure(figsize=(10, 6))
df.groupby(["Category", "Size"]).size().unstack().plot(kind="bar")
plt.xlabel("Category")
plt.ylabel("Number of Orders")
plt.title("Size Distribution Across Categories")
plt.show()
```

#### Note: Different product categories show different size preferences among customers

```python
# Get top 10 shipping states by order count
top_10_states = df["ship-state"].value_counts().head(10)

# Filter dataframe for top 10 states
filtered_df = df[df["ship-state"].isin(top_10_states.index)]

# Create the plot
plt.figure(figsize=(12, 6))
sns.countplot(
    data=filtered_df,
    x="ship-state",
    order=top_10_states.index
)

# Labels and title
plt.xlabel("Shipping State")
plt.ylabel("Number of Orders")
plt.title("Top 10 Shipping States by Order Count")

# Improve readability
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

#### Note: The graph clearly indicates that Maharashtra has the highest number of buyers among all shipping states.

## Conclusion 

 #### The analysis of Amazon sales data reveals that customer demand is concentrated in specific regions, product sizes, and categories, with Maharashtra emerging as the top shipping state, M-size being the most purchased, and retail customers dominating overall sales, providing actionable insights for inventory planning and targeted marketing strategies.
